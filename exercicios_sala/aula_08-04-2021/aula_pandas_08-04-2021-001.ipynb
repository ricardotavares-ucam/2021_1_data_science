```python
df.drop(['city'], axis = 1, inplace = True)
```

```python
df.drop(df.loc[df['state'] == 'TOTAL'].index, inplace = True)
```

```python
df_sum_by_state = df.groupby(by = 'state')['newCases', 'newDeaths'].sum().reset_index()
```

```python
df_sum_by_state.head()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plt.rcParams['font.family'] = 'Arial'
sns.set_style(style='white')
sns.barplot(ax = ax, x = 'state', y = 'newDeaths', 
    data = df_sum_by_state.sort_values('newDeaths', ascending = False), alpha=0.8, palette='Blues_d')
fig.suptitle('Números de mortes por Coronavirus por Estado', fontsize = 16, fontweight = 'bold')
ax.set_xlabel('Estado', fontsize = 12)
ax.set_ylabel('Número de mortes', fontsize = 12)
ax.set(yticks=[])
sns.despine(offset=10, trim=True);

ax.set_ylim([df_sum_by_state['newDeaths'].min(),df_sum_by_state['newDeaths'].max()*1.2])
for i, v in enumerate(df_sum_by_state['newDeaths'].sort_values(ascending = False)):
    ax.text(i, v + 3, str(v), rotation=45)
plt.show(ax)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
plt.rcParams['font.family'] = 'Arial'
sns.set_style(style='white')
sns.barplot(ax = ax, x = 'state', y = 'newCases', 
    data = df_sum_by_state.sort_values('newCases', ascending = False), alpha=0.8, palette='Blues_d')
fig.suptitle('Números de Casos de Coronavirus por Estado', fontsize = 16, fontweight = 'bold')
ax.set_xlabel('Estado', fontsize = 12)
ax.set_ylabel('Número de casos', fontsize = 12)
ax.set(yticks=[])
sns.despine(offset=10, trim=True);

ax.set_ylim([df_sum_by_state['newCases'].min(),df_sum_by_state['newCases'].max()*1.2])
for i, v in enumerate(df_sum_by_state['newCases'].sort_values(ascending = False)):
    ax.text(i, v + 3, str(v), rotation=45)
plt.show(ax)
```

```python
df_sum_by_state['tx_mortalidade'] = df_sum_by_state['newDeaths']/df_sum_by_state['newCases']
```

```python
media_geral = df_sum_by_state['newDeaths'].sum()/df_sum_by_state['newCases'].sum()
media_geral
```

```python
fig, ax = plt.subplots(figsize=(10,5))
sns.set_style(style='white')
sns.barplot(ax = ax, x = 'state', y = 'tx_mortalidade', 
    data = df_sum_by_state.sort_values('tx_mortalidade', ascending = False), alpha=0.8, palette='Blues_d')
sns.lineplot(ax = ax, x = 'state', y = media_geral, 
    data = df_sum_by_state.sort_values('tx_mortalidade', ascending = False), linestyle='dashed', color='orange')
fig.suptitle('Taxa de mortalidade do Coronavirus por Estado', fontsize = 16, fontweight = 'bold')
ax.set_xlabel('Estado', fontsize = 12)
ax.set_ylabel('Taxa de mortalidade', fontsize = 12)
ax.set(yticks=[])
for i, v in enumerate(df_sum_by_state['tx_mortalidade'].sort_values(ascending = False)):
    ax.text(i, v*1.025, str(round(v,2)))
sns.despine(offset=10, trim=True);
plt.show(ax)
```
